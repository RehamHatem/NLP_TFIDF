```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from transformers import pipeline

def generate_documents(phrases, fields, model_name='EleutherAI/gpt-neo-2.7B'):
    # Initialize text generation pipeline
    gen = pipeline('text-generation', model=model_name)

    # Dictionary to store generated documents
    documents = {}
    unique_words = set()

    for field in fields:
        # Generate a document for each field
        documents[field] = []
        for phrase in phrases:
            # Generate text based on the phrase for the current field
            generated_text = gen(phrase, max_length=50, num_return_sequences=1)[0]['generated_text']
            # Preprocess the generated text
            preprocessed_text = preprocess_text(generated_text)
            documents[field].append(preprocessed_text)

            # Extract unique words
            tokens = word_tokenize(preprocessed_text)
            unique_words.update(tokens)

    return documents, unique_words

def preprocess_text(text):
    # Remove non-alphanumeric characters and extra whitespace
    cleaned_text = re.sub(r'[^\w\s]', '', text)
    # Normalize to lowercase
    cleaned_text = cleaned_text.lower()
    return cleaned_text

def clean_text(text):
    # Remove non-alphanumeric characters and extra whitespace
    cleaned_text = re.sub(r'[^\w\s]', '', text)
    # Normalize to lowercase
    cleaned_text = cleaned_text.lower()
    return cleaned_text

def tokenize_text(text):
    # Tokenize the text into words
    tokens = text.split()
    return tokens

def lemmatize_text(tokens):
    # Initialize WordNet Lemmatizer
    lemmatizer = WordNetLemmatizer()
    # Lemmatize each word, excluding words with less than 2 characters
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens if len(token) > 3]
    return lemmatized_tokens

def remove_stopwords(tokens):
    # Get English stop words
    stop_words = set(stopwords.words('english'))
    # Remove stop words
    filtered_tokens = [token for token in tokens if token not in stop_words]
    return filtered_tokens

def get_unique_words(documents):
    # Initialize an empty set to store unique words
    unique_words = set()
    # Iterate through each document
    for document in documents:
        cleaned_paragraph = clean_text(document)
        tokens = tokenize_text(cleaned_paragraph)
        lemmatized_tokens = lemmatize_text(tokens)
        filtered_tokens = remove_stopwords(lemmatized_tokens)
        unique_words.update(filtered_tokens)
    return unique_words

def get_corpus(documents):
    # Clean, tokenize, lemmatize, and remove stopwords from each document
    corpus = [' '.join(remove_stopwords(lemmatize_text(tokenize_text(clean_text(doc))))) for doc in documents]
    return corpus

# Example phrases and fields
phrases = ["Introduction to AI", "Future of technology", "Climate change impacts"]
fields = ["Body"]

# Generate documents
generated_documents, unique_words = generate_documents(phrases, fields)

# Print generated documents
for field, documents in generated_documents.items():
    print(f"{field}:")
    for i, doc in enumerate(documents, start=1):
        print(f"Document {i}: {doc}\n")

# Further processing
corpus = []
for field, documents in generated_documents.items():
    corpus.extend(documents)

# Clean, tokenize, lemmatize, and remove stopwords from each document
corpus = get_corpus(corpus)

print("\nProcessing on data: Clean, tokenize, lemmatize, and remove stopwords from each document\n")
print("Corpus:\n", corpus)
```

```python
unique_words_generated = get_unique_words(corpus)

# Print unique words
print("\nUnique words in the generated documents:\n")
print(unique_words_generated)
```

**built in (sklearn)**

*   1. Get TF for each word for all documents.
*   2. Get IDF for each word.

*   3. Multiply TF * IDF
*  4. Get Normalized TFIDF

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Step 1: Calculate TF for each word for all documents
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(corpus)

# Step 2: Calculate IDF for each word
idf = vectorizer.idf_

# Step 3: Multiply TF * IDF
tfidf_scores = tfidf_matrix.toarray() * idf

# Step 4: Normalize TF-IDF
# Normalization to unit norm (vector magnitude becomes 1)
normalized_tfidf_scores = tfidf_scores / (tfidf_scores**2).sum(axis=1, keepdims=True)**0.5

# Step 5: Print the results
print("TF-IDF scores for all documents:")
print(pd.DataFrame(normalized_tfidf_scores, columns=vectorizer.get_feature_names_out()))
```

**From Scratch**

*   1. Get TF for each word for all documents.
*   2. Get IDF for each word.

*   3. Multiply TF * IDF
*  4. Get Normalized TFIDF



```python
import math
from collections import Counter

# Step 1: Calculate TF for each word for all documents
def calculate_tf(document):
    word_counts = Counter(document)
    total_words = len(document)
    tf = {word: count / total_words for word, count in word_counts.items()}
    return tf

# Step 2: Calculate IDF for each word
def calculate_idf(documents):
    total_documents = len(documents)
    idf = {}
    for document in documents:
        for word in document:
            if word not in idf:
                # Count how many documents contain the word
                document_frequency = sum([1 for doc in documents if word in doc])
                # Calculate IDF
                idf[word] = math.log((total_documents+1) / (1 + document_frequency))+1
    return idf

# Step 3: Multiply TF * IDF
def calculate_tfidf(tf, idf):
    tfidf = {word: tf_value * idf.get(word, 0) for word, tf_value in tf.items()}
    return tfidf

# Step 4: Get Normalized TFIDF
def normalize_tfidf(tfidf):
    total = math.sqrt(sum(value ** 2 for value in tfidf.values()))
    normalized_tfidf = {word: value / total for word, value in tfidf.items()}
    return normalized_tfidf

# Combine all documents
all_documents = []
for documents in generated_documents.values():
    all_documents.extend(documents)

# Tokenize, lemmatize, and remove stopwords for all documents
tokenized_documents = [remove_stopwords(lemmatize_text(tokenize_text(clean_text(doc)))) for doc in all_documents]

# Step 1: Calculate TF for all documents
tf_all_documents = [calculate_tf(doc) for doc in tokenized_documents]

# Step 2: Calculate IDF for all words
idf_all_words = calculate_idf(tokenized_documents)

# Step 3: Multiply TF * IDF
tfidf_all_documents = [calculate_tfidf(tf, idf_all_words) for tf in tf_all_documents]

# Step 4: Get Normalized TFIDF
normalized_tfidf_all_documents = [normalize_tfidf(tfidf) for tfidf in tfidf_all_documents]

# Print the results
print("Normalized TF-IDF scores for all documents:")
for i, tfidf in enumerate(normalized_tfidf_all_documents):
    print(f"Document {i+1}:")
    for word, score in tfidf.items():
        print(f"{word}: {score:.4f}")
    print()
```
